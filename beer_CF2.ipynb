## 인덱스가 beer_name 인걸로 사용
beer_name - ID 피벗테이블
table1 = a.pivot_table(a, index = ['beer_name'], columns = ['ID'])

```python
# 아이템기반 협업 필터링을 위해 인덱스가 beer_name 인걸로 사용

# 피벗테이블 NaN처리
table1.fillna(0,inplace=True)
table1.head()
```

# 아이템기반 협업 필터링

## 유사한 아이템끼리 추천을 해주는 방식 
평점이 비슷한 맥주를 추천해 주는 것

코사인 유사도 값을 이용하여 유사도 계산

```python
# 유사도값 추출 (코사인 유사도)

from sklearn.metrics.pairwise import cosine_similarity

similarity_rate = cosine_similarity(table1, table1)
print(similarity_rate)
```

```python
# 유사도값을 가진 데이터프레임 새성
simlarity_rate_df = pd.DataFrame(
    data = similarity_rate,
    index = table1.index,
    columns = table1.index)
```

```python
simlarity_rate_df.head()
```

```python
simlarity_rate_df.drop(['0.0'], axis = 0)
```

# 맥주추천 사용자 함수 생성
사용자가 어떤 맥주를 선택하였을 때(매개변수로 맥주이름 입력)

비슷한 맥주를 추천

```python
# 가장 유사도가 높은 top5
def recommand_beer(beer_name) : 
    return simlarity_rate_df[beer_name].sort_values(ascending=False)[:6]
recommand_beer("Tiger Beer")
```
