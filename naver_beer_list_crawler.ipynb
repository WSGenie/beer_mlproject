```python
from tqdm import tqdm
```

```python
# url 설정 (페이지 넘어가는 url 설정을 어떻게 할건지?)
# 순환시키면서 저 requests 요청을 어떻게 컨트롤 할 건지?
beer_list = []

for page in tqdm(range(79)) : 
    
    seed = np.random.randint(100)
    np.random.seed(seed)
    a = np.random.randint(5)
    time.sleep(a)
    
    # url 설정
    url = f'https://terms.naver.com/list.naver?cid=48273&categoryId=48273&page={page}'
    print(f'{page}페이지 크롤링 중입니다.')
   
    # requests로 데이터 요청하기
    resp = requests.get(url)
    
    # html로 변환
    html = BeautifulSoup(resp.text, 'html.parser')
    
    #데이터 선별
    for item in html.select('strong.title')[4: ] :
        beer_list.append(item.text)

#beer_array = np.array(beer_list)

print ('크롤링 완료!')
```
